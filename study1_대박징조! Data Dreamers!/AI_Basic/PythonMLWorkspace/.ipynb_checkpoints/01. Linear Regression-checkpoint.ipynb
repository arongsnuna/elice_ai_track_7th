```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('LinearRegressionData.csv')
```

```python
X = dataset.iloc[:,:-1].values #      ( - ) -  
y = dataset.iloc[:,-1].values #    ( - ) -  
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0) # 80  20
```

###    

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train) # 
```

###   ()

```python
plt.scatter(X_train,y_train,color='blue') #
plt.plot(X_train,reg.predict(X_train), color='green') #
plt.title('Score by hours(train data)')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

###   ()

```python
plt.scatter(X_test,y_test,color='blue') #
plt.plot(X_train,reg.predict(X_train), color='green') #
plt.title('Score by hours(test data)')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
reg.coef_ #
```

```python
reg.intercept_ #y
```

###  

```python
reg.score(X_test, y_test) #    
```

```python
reg.score(X_train, y_train) #    
```

   Stochastic Gradient Descent

          

```python
from sklearn.linear_model import SGDRegressor #SGD: Stochastic Gradient Descent   
sr = SGDRegressor()
sr.fit(X_train, y_train)
```

```python
plt.scatter(X_train,y_train,color='blue') #
plt.plot(X_train,sr.predict(X_train), color='green') #
plt.title('Score by hours(test data, SGD)')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
sr.coef_ # 10.4916
```

```python
sr.intercept_ #y 0.6115
```

```python
sr.score(X_test, y_test) #    
```

```python
sr.score(X_train, y_train) #    
```

```python
sr2 = SGDRegressor(max_iter=1000, eta0=1e-4, random_state=0)
# (  ): max_iter
# : eta0, 
#verbose=1
```

```python
sr2.fit(X_train, y_train)
```

```python
plt.scatter(X_train,y_train,color='blue') #
plt.plot(X_train,sr2.predict(X_train), color='green') #
plt.title('Score by hours(test data, SGD)')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```
